```python
from scipy import integrate
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# integrate 对函数 (1 -x^2)^(1/2) 进行积分
# x^2 + y^2 =1 是一个圆
# s = pi * r**2 
# 1. 首先画一个圆, 圆的上半部分
c = lambda x : (1- x**2)**(1/2)
x = np.linspace(-1, 1, 100)
y = c(x)
plt.figure(figsize=(5,5)) # 调整比例
plt.plot(x, y)
plt.plot(x, -y)
```

```python
# s = pi * r**2
# pi = s / (r**2)
# # r = 1
# pi = s
# 可以求一个半圆面积 * 2 就是 pi
# quad 是定积分
integrate.quad(c, -1, 1)[0] * 2
```
